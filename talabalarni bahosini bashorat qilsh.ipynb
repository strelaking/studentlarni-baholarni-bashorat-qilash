Ma'lumotlar to'plami foydalanishga tayyor, chunki ma'lumotlarda null qiymatlar mavjud emas. Ma'lumotlarda talabalar tanlagan kurslar soni haqidagi ma'lumotlarni o'z ichiga olgan ustun mavjud. Keling, ushbu ustunning barcha qiymatlari qiymatlari sonini ko'rib chiqaylik:

Shunday qilib, talabalar kamida uchta va maksimal sakkizta kursni tanladilar. Kurslar soni talaba baholariga ta'sir qiladimi yoki yo'qligini bilish uchun tarqalish sxemasini ko'rib chiqamiz:

Yuqoridagi ma'lumotlar vizualizatsiyasiga ko'ra, agar talaba har kuni ko'proq vaqt o'qiyotgan bo'lsa, kurslar soni talabaning baholariga ta'sir qilmasligi mumkinligini aytishimiz mumkin. Keling, har kuni o'qigan vaqt va talaba tomonidan to'plangan ball o'rtasidagi bog'liqlikni ko'rib chiqaylik:

```python
correlation = data.corr()
print(correlation["Marks"].sort_values(ascending=False))
```

O'rganilgan vaqt va olingan belgilar o'rtasida chiziqli bog'liqlik mavjudligini ko'rishingiz mumkin. Bu shuni anglatadiki, talabalar o'qishga qanchalik ko'p vaqt sarflasa, ular shunchalik yaxshi ball oladi.

Keling, talabalar tomonidan to'plangan ballar va ma'lumotlarning qolgan ikkita ustuni o'rtasidagi bog'liqlikni ko'rib chiqaylik:

```python
x = np.array(data[["time_study", "number_courses"]])
y = np.array(data["Marks"])
xtrain, xtest, ytrain, ytest = train_test_split(x, y, 
                                                test_size=0.2, 
                                                random_state=42)
```

Shunday qilib, time_studied ustuni boshqa ustunga qaraganda belgilar ustuni bilan ko'proq bog'langan.

Talaba belgilarini bashorat qilish modeli
Endi talaba baholarini bashorat qilish uchun mashinani o'rganish modelini o'rgatish vazifasiga o'tamiz. Bu erda men birinchi navbatda ma'lumotlarni o'quv va test to'plamlariga bo'lishdan boshlayman:

```python
model = LinearRegression()
model.fit(xtrain, ytrain)
model.score(xtest, ytest)
```

Endi men chiziqli regressiya algoritmidan foydalangan holda mashinani o'rganish modelini o'rgataman:

```python
# Features = [["time_study", "number_courses"]]
features = np.array([[4.508, 3]])
model.predict(features)
```
